```python
# ðŸ§  Step 1: Import required libraries
import os
from dotenv import load_dotenv
from agents import Agent, Runner
from agents.extensions.visualization import draw_graph
from IPython.display import display

# âœ… Step 2: Load environment variables
load_dotenv()
api_key = os.getenv("OPENAI_API_KEY")
if not api_key:
    raise ValueError("OPENAI_API_KEY is not set.")

# âœ… Step 3: Create an async function to run the agent
import nest_asyncio
import asyncio
#nest_asyncio.apply()  # ðŸ‘ˆ this is required to allow asyncio.run() inside Jupyter

async def run_agent():
    agent1 = Agent(
        name="Assistant",
        instructions="You are a helpful assistant"
    )
    result = await Runner.run(agent1, "Write a haiku about recursion in programming.")
    print(result.final_output)
    display(draw_graph(agent1))


# âœ… Step 4: Run the async agent function
await run_agent()
```
